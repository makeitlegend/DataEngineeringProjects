###Column cross-checking

```python
#'formula_pretty', 'n_elements' and 'elements' columns have a direct and visible relationship
#I want to cross-check 'formula_pretty' with 'n_elements' first

#make a copy of the dataset
df = df_materials.copy()

'''
extract chemical element symbols directly from 'formula_pretty'
I use a regex that captures valid element symbols:
-an uppercase letter
-followed optionally by a lowercase letter
Examples:
"LiMg"        -> ["Li", "Mg"]
"LiFe2F5"     -> ["Li", "Fe", "F"]
"Li(Fe2O3)4"  -> ["Li", "Fe", "O"]
"LiMn2O4"     -> ["Li", "Mn", "O"]
Note: re.findall returns all matches in order of appearance
'''

df["elements_extracted"] = df["formula_pretty"].apply(
    lambda f: re.findall(r"[A-Z][a-z]?", f)
)

'''
count how many unique elements were detected
I use a set() because formulas can repeat elements
Example: Fe2O3 → Fe, O repeated multiple times in the formula
'''

df["n_elements_extracted"] = df["elements_extracted"].apply(
    lambda elems: len(set(elems))
)

#compare extracted element count with the column 'n_elements'
#this gives a final True/False column indicating consistency
df["elements_match"] = df["n_elements_extracted"] == df["n_elements"]

df["elements_match"].value_counts()
```

```python
#I want to cross-check 'n_elements' with 'elements' next

'''
count how many elements are in the 'elements' column
I use string operations to count elements separated by commas
Examples:
"['Li', 'O']"          -> 2
"['Fe', 'Li', 'O']"    -> 3
"['Co', 'Cr', 'Fe']"   -> 3
Note: this works because each element is separated by a comma
'''

#remove brackets and spaces, then split by comma
df_materials["n_elements_from_elements"] = df_materials["elements"].apply(
    lambda s: len(s.strip("[]").replace("'", "").replace(" ", "").split(",")) if pd.notnull(s) else 0
)


#compare extracted element count with the column 'n_elements'
#this gives a final True/False column indicating consistency
df_materials["n_elements_match"] = df_materials["n_elements_from_elements"] == df_materials["n_elements"]

df_materials["n_elements_match"].value_counts()
```

```python
#finally, I want to cross-check 'formula_pretty' with 'elements'

'''
I use the extracted counts from both columns
True means the number of elements detected in 'formula_pretty'
matches the number of elements listed in 'elements'
'''

df_materials["formula_vs_elements_match"] = (
    df["n_elements_extracted"] == df_materials["n_elements_from_elements"]
)

df_materials["formula_vs_elements_match"].value_counts()
```
