##Data Cleaning

```python
#material_id column -> common values: mp-20150, mp-1193296, mp-2659

#Missing values-material_id column: 0
#Duplicated values-material_id column: 57492
#Total values-material_id column: 79239

import re

#convert material_id to string, strip of white spaces
df_materials['material_id'] = df_materials['material_id'].astype(str).str.strip()

#sometimes CSV or Excel imports add invisible characters that strip() won’t catch
#this keeps only standard ASCII characters
df_materials['material_id'] = df_materials['material_id'].str.replace(r"[^\x00-\x7F]+", "", regex=True)

#make sure all values converted to str
df_materials['material_id'].map(type).unique()

#show top 10 longest values
#sorted_values = sorted(
    #df_materials['material_id'].map(str).dropna().unique(),
    #key=len,
    #reverse=True
#)
#sorted_values[:10]

#regex pattern for 'mp-' + 3 to 7 digits
pattern_strict = r'^mp-(\d{3}|\d{4}|\d{5}|\d{6}|\d{7})$'

#create mask
mask_valid = df_materials['material_id'].str.match(pattern_strict)

#count valid vs invalid
valid_count = mask_valid.sum()
invalid_count = (~mask_valid).sum()

print(f"Valid material_id values: {valid_count}")
print(f"Invalid material_id values: {invalid_count}")

#display invalid ones
df_materials.loc[~mask_valid, 'material_id'].value_counts()

len(df_materials)
```

```python
#formula_pretty column -> common values: LiMg, LiFe2F5, Li(Fe2O3)4, LiMn2O4

#Missing values-formula_pretty column: 0
#Duplicated values-formula_pretty column: 77465
#Total values-formula_pretty column: 86988

#convert formula_pretty to string, strip of white spaces
df_materials['formula_pretty'] = df_materials['formula_pretty'].astype(str).str.strip()

#sometimes CSV or Excel imports add invisible characters that strip() won’t catch
#this keeps only standard ASCII characters
df_materials['formula_pretty'] = df_materials['formula_pretty'].str.replace(r"[^\x00-\x7F]+", "", regex=True)

#make sure all values converted to str
df_materials['formula_pretty'].map(type).unique()

#regex pattern (all 118 IUPAC elements included)
pattern_formula = (
    r'^('
        r'('
            r'H|He|Li|Be|B|C|N|O|F|Ne|'
            r'Na|Mg|Al|Si|P|S|Cl|Ar|'
            r'K|Ca|Sc|Ti|V|Cr|Mn|Fe|Co|Ni|Cu|Zn|Ga|Ge|As|Se|Br|Kr|'
            r'Rb|Sr|Y|Zr|Nb|Mo|Tc|Ru|Rh|Pd|Ag|Cd|In|Sn|Sb|Te|I|Xe|'
            r'Cs|Ba|La|Ce|Pr|Nd|Pm|Sm|Eu|Gd|Tb|Dy|Ho|Er|Tm|Yb|Lu|'
            r'Hf|Ta|W|Re|Os|Ir|Pt|Au|Hg|Tl|Pb|Bi|Po|At|Rn|'
            r'Fr|Ra|Ac|Th|Pa|U|Np|Pu|Am|Cm|Bk|Cf|Es|Fm|Md|No|Lr|'
            r'Rf|Db|Sg|Bh|Hs|Mt|Ds|Rg|Cn|Nh|Fl|Mc|Lv|Ts|Og'
        r')'
        r'\d*'
        r'|\('
            r'[A-Za-z0-9]+'
        r'\)\d*'
    r')+$'
)

#create mask
mask_valid = df_materials['formula_pretty'].str.match(pattern_formula)

#count valid vs invalid
valid_count = mask_valid.sum()
invalid_count = (~mask_valid).sum()

print(f"Valid formula_pretty values: {valid_count}")
print(f"Invalid formula_pretty values: {invalid_count}")

#display invalid ones
df_materials.loc[~mask_valid, 'formula_pretty'].value_counts()

len(df_materials)
```
