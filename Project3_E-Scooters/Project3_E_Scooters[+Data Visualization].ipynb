##Data Cleaning

```python
#index column -> common values: from 0 to 61499

#Missing values-index column: 0
#Duplicated values-index column: 0
#Total values-index column: 61500

#this column has dtype int64, which means all values are integers

#range check
df_e_scooters['index'].min(), df_e_scooters['index'].max()

#confirm first and last value
#df_e_scooters['index'].iloc[0], df_e_scooters['index'].iloc[-1]

#confirm values are incrementing
df_e_scooters['index'].is_monotonic_increasing

#confirm range
expected = pd.Series(range(61500))
df_e_scooters['index'].sort_values(ignore_index=True).equals(expected)
```

```python
#timestamp column -> common value: 2022-10-29 15:19:02.251

#Missing values-timestamp column: 0
#Duplicated values-timestamp column: 0
#Total values-timestamp column: 61500

#convert to datetime and keep the format
df_e_scooters["timestamp"] = pd.to_datetime(
    df_e_scooters["timestamp"],
    format="%Y-%m-%d %H:%M:%S.%f"
)

#confirm conversion
df_e_scooters["timestamp"].dtype

#if any row doesn't match the format, this will produce NaT when errors="coerce" is used
tmp = pd.to_datetime(
    df_e_scooters["timestamp"],
    format="%Y-%m-%d %H:%M:%S.%f",
    errors="coerce"
)

#confirm invalid rows
invalid = df_e_scooters[tmp.isna()]
print(invalid)

#confirm only one date is present in the rows
df_e_scooters["timestamp"].dt.date.nunique()
#-> output is "1", so the date part (YYYY-MM-DD) is always 2022-10-29

#confirm min and max values
df_e_scooters["timestamp"].min(), df_e_scooters["timestamp"].max()

#confirm the time range (→ ~10 minutes and 16 seconds)
df_e_scooters["timestamp"].max() - df_e_scooters["timestamp"].min()
```

```python
#acc_x column -> common values: 2.078125, -1.296875

#Missing values-acc_x column: 0
#Duplicated values-acc_x column: 60967
#Total values-acc_x column: 61500

#dtype is float64 (no whitespaces or other characters besides '-')

#confirm min and max values
df_e_scooters["acc_x"].agg(["min", "max"])

#confirm how many values are 0 (75 values)
(df_e_scooters["acc_x"] == 0).sum()

#check how many values have X (<=6) number of decimals
decimal_counts = (
    df_e_scooters["acc_x"]
    .astype(str)
    .str.split(".")
    .str[1]
    .str.len()
    .value_counts()
    .sort_index()
)

decimal_counts

#confirm how many values are positive (59303) vs negative (2122)
df_e_scooters["acc_x"].agg(
    positive=lambda x: (x > 0).sum(),
    negative=lambda x: (x < 0).sum()
)

#59303 (positive values) + 2122 (negative values) + 75 ('0' values) = 61500 (total values)
```

```python
#time column -> common values: 0, 0.01, 0.061, 42.158, 213.178

#Missing values-time column: 0
#Duplicated values-time column: 0
#Total values-time column: 61500

#confirm min (0.000) and max (616.379) values
df_e_scooters["time"].agg(["min", "max"])

#confirm how many values are 0 (1 value)
(df_e_scooters["time"] == 0).sum()

#check how many values have X (<=3) number of decimals
decimal_counts = (
    df_e_scooters["time"]
    .astype(str)
    .str.split(".")
    .str[1]
    .str.len()
    .value_counts()
    .sort_index()
)

decimal_counts #173-1 dec, 1537-2 dec, 59790-3 dec

#confirm how many values are positive (61499) vs negative (0)
df_e_scooters["time"].agg(
    positive=lambda x: (x > 0).sum(),
    negative=lambda x: (x < 0).sum()
)

#confirm that the order of the values is ascending
is_sorted = df_e_scooters["time"].is_monotonic_increasing
is_sorted

#61499 (positive values) + 0 (negative values) + 1 ('0' values) = 61500 (total values)
```

```python
#label column -> common values: standing, tarmac, tile-stone, stone, grass, sync

#Missing values-label column: 0
#Duplicated values-label column: 61494
#Total values-label column: 61500

import re

#convert label to string, strip of white spaces
df_e_scooters['label'] = df_e_scooters['label'].astype(str).str.strip()

#sometimes CSV or Excel imports add invisible characters that strip() won’t catch
#this keeps only standard ASCII characters
df_e_scooters['label'] = df_e_scooters['label'].str.replace(r"[^\x00-\x7F]+", "", regex=True)

#make sure all values converted to str
df_e_scooters['label'].map(type).unique()

#check most common values
df_e_scooters['label'].value_counts()

#create regex pattern for the 6 known labels
pattern_labels = r'^(standing|tarmac|tile-stone|stone|grass|sync)$'

#create mask of valid labels
mask_valid_labels = df_e_scooters['label'].str.match(pattern_labels)

#count valid vs invalid
valid_count = mask_valid_labels.sum()
invalid_count = (~mask_valid_labels).sum()

print(f"Valid labels: {valid_count}")
print(f"Invalid labels: {invalid_count}")

#display invalid labels
#df_e_scooters.loc[~mask_valid_labels, 'label'].value_counts()
```

```python
#sample column

#Missing values-sample column: 0
#Duplicated values-sample column: 0
#Total values-sample column: 61500

#this column has dtype int64, which means all values are integers

#confirm min (0) and max (61499) values
df_e_scooters["sample"].agg(["min", "max"])

#confirm first (0) and last value (61499)
df_e_scooters['sample'].iloc[0], df_e_scooters['sample'].iloc[-1]

#confirm that the order of the values is ascending
is_sorted = df_e_scooters["sample"].is_monotonic_increasing
is_sorted

#confirm range
expected = pd.Series(range(61500))
df_e_scooters['sample'].sort_values(ignore_index=True).equals(expected)
```
