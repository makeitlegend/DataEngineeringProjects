##Data Exploration

```python
df_e_scooters = pd.read_csv("/content/E-ScootersData.csv")
```

```python
#index column
missing_values_df_e_scooters_index = df_e_scooters['index'].isna().sum()
duplicated_values_df_e_scooters_index = df_e_scooters['index'].duplicated().sum()
total_values_df_e_scooters_index = df_e_scooters['index'].count()

print(f'Missing values-index column: {missing_values_df_e_scooters_index}')
print(f'Duplicated values-index column: {duplicated_values_df_e_scooters_index}')
print(f'Total values-index column: {total_values_df_e_scooters_index}')
```
